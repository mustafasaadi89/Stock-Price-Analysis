```python
# Dependencies
import requests
import json
import matplotlib.pyplot as plt
```

```python
# possible ranges: 5y, 2y, 1y, ytd, 6m, 3m, 1m, 1d
symbol = 'fb'
the_range = '6m'

url = "https://api.iextrading.com/1.0/stock/" + symbol + "/batch?types=quote,news,chart&range="+the_range+"&last=5"
url2 = "https://api.iextrading.com/1.0/stock/" + symbol + "/chart/"+the_range
```

```python
# Print the response object to the console
print(requests.get(url))
```

```python
# Retrieving data and converting it into JSON
print(requests.get(url).json())
```

```python
# Pretty Print the output of the JSON
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
# print the dates and closing prices for the stock over the past the_range
close = []
date = []
for x in response["chart"]:
    close.append(x["close"])
    date.append(x["date"])
print(close)
```

```python
plt.plot(date, close)
plt.show()
```

```python
#print the headlines of the last 5 articles about the stock
for x in response["news"]:
    print(x["headline"])
```
